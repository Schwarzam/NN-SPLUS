```
from IPython.display import clear_output
import pandas as pd
from PIL import Image
```

```
esc = input("reset? ")
if esc == 'y':
  newdf = pd.DataFrame(columns= ['ID', 'Class'])
  print('done!')
```

```
path = "/Users/oliveira/Documents/Galaxies.csv"
df = pd.read_csv(path)
```

```
esc = input("load df?  ")
if esc == 'y':
  newdf = pd.read_csv("/Users/oliveira/Documents/Classification.csv")
  del newdf['Unnamed: 0']
  print(newdf)
  print('done')
```

```
start_p = len(newdf)
for key, value in df[start_p:].iterrows():
  try:
    print('Galaxy number: ', key, 'of', len(df), '\nID: ', value['ID'], '\nRA and Dec: ', value['ra'], value['dec'] )

    im = Image.open(f"/Users/oliveira/Documents/Galaxies/samples/{value['ID']}.png")
    im = im.resize((512,512), Image.ANTIALIAS)
    display(im)

    print("1 Elliptical \n2 Lenticular \n3 Spiral \n4 Merger (Two objects) \n5 Merger (Assimetrica) \n6 Dont use img \n0 to EXIT")

    num = int(input("What is the img class:  \n"))

    if num > 6 or num < 1:
      if num == 0:
        break
      else:
        print('invalido')
        break
    if num == 1:
      num = 'Elliptical'
    if num == 2:
      num = 'Lenticular' 
    if num == 3:
      num = 'Spiral'
    if num == 4:
      num = 'MergerObj' 
    if num == 5:
      num = 'MergerAs'
    if num == 6:
      num = 'Leave'

    classification = {'ID': [value['ID']],
          'Class': [num]}

    part = pd.DataFrame(classification, columns= ['ID', 'Class'])
    newdf = newdf.append(part)
  except:
    pass

  #  z = int(input('continuar: '))
  #  print(z)
  #  if z == 0: 
  #    break

  clear_output(wait=True)
```

```
prog = input("save progress? ")
if prog == 'y':
  newdf.to_csv("/Users/oliveira/Documents/Classification.csv")
  print('done!')
```
