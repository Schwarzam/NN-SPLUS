```python
new_col = pd.Series(range(7050))
df['Class'] = new_col
df['P_bulge'] = new_col
df['P_merg'] = new_col
df['P_disk'] = new_col
df['P_S0'] = new_col

arr = np.array([])
ra2 = df1.RA_1.to_numpy()
dec2 = df1.dec_1.to_numpy()
id2 = df1.ID.to_numpy()


idclass = df_class.ID.to_numpy()
for key, value in df.iterrows():
    ra = value.RA
    dec = value.DEC
    
    c = SkyCoord(ra=ra*u.degree, dec=dec*u.degree)
    catalog = SkyCoord(ra=ra2*u.degree, dec=dec2*u.degree)
    idx, d2d, d3d = c.match_to_catalog_sky(catalog)
    
    if (d2d < 1*u.arcsec):
        if id2[idx] in idclass:
            arr = np.append(arr, value.ID)

        obj = df_class.loc[df_class['ID'] == id2[idx], 'Class']
        if len(obj) > 0:
            df.loc[df['ID'] == value.ID, 'Class'] = obj.values[0]
        
        obj = df1.loc[df1['ID'] == id2[idx], 'P_bulge']
        if len(obj) > 0:
            df.loc[df['ID'] == value.ID, 'P_bulge'] = obj.values[0]

        obj = df1.loc[df1['ID'] == id2[idx], 'P_merg']
        if len(obj) > 0:
            df.loc[df['ID'] == value.ID, 'P_merg'] = obj.values[0]

        obj = df1.loc[df1['ID'] == id2[idx], 'P_disk']
        if len(obj) > 0:
            df.loc[df['ID'] == value.ID, 'P_disk'] = obj.values[0]

        obj = df1.loc[df1['ID'] == id2[idx], 'P_S0']
        if len(obj) > 0:
            df.loc[df['ID'] == value.ID, 'P_S0'] = obj.values[0]
```
